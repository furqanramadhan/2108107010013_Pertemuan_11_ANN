```python
# import library yang dibutuhkan

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset
df = pd.read_csv('dataset/megabagus/Data_bank_churn.csv')
df.head(5)
```

```python
# Summary statistics dataset in numerical variables
round(df.describe(), 2)
```

```python
df.info
```

```python
""" Slicing X and y variables
X, variabel independen, membutuhkan kolom 3-12
y, variabel terikat, membutuhkan kolom terakhir
"""
X = df.iloc[:, 3:13].values
y = df.iloc[:, 13].values
```

```python
# Convert to categorical to numeric in Gender
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

labelencoder = LabelEncoder()
X[:,2] = labelencoder.fit_transform(X[:,2])
```

```python
# Create dummy variable in Geography
# columnTransformer berfungsi untuk memberikan nama pada negara menjadi numeric, 0..1..2, dst, tetapi tidak membuat nilai dari kolom tersebut 
# ex : 0 [France] tidak lebih kecil dari 1 [Spain]
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(columnTransformer.fit_transform(X), dtype = np.float64)

# remove 1 column dummy variabel in Geography
X = X[:,1:]
```

```python
# Memasukkan data kedalam training dan test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Import library TensorFlow(2.0), Keras, Sequential, dan Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initialize ANN
MesinKlasifikasi = Sequential()
```

```python
# Menambah input layer dan hidden layer pertama
MesinKlasifikasi.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
# Menambah hidden layer kedua
MesinKlasifikasi.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
# Menambah output layer
MesinKlasifikasi.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Running ANN
MesinKlasifikasi.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting ANN to training set 
MesinKlasifikasi.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Memprediksi hasil test set
y_pred = MesinKlasifikasi.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Membuat confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
